## 2. 変数 
変数はデータを格納するための箱。  
変数を用意することで、変数に値を代入したり、変数から値を出力することができる。  

※ プログラムにおける「=」は等しいではなく代入という意味

```python
# 変数my_varに100を「代入」
my_var = 100
print(my_var)
```

```python
# 変数my_varに200を「代入」　
# 変数my_varの中身が100→200に上書き
my_var = 200
print(my_var)
```

## 3. リスト 
### 3.1. リストの概要
リストとは要素を複数格納することができるもの。  
リストの要素には数値や文字列など Pythonで扱えるオブジェクトであれば何でも指定することができる。  


```python
# 要素として数値を格納するリスト
[1, 2, 3, 4, 5, 6, 7]
```

```python
# 変数を格納することも可能
[num_01, num_02, num_03]
```

```python
# 同じリストで異なる型の要素を格納することが可能
[1, "dog", 12.424]
```

```python
# 空のリストを生成することも可能
[]
```

生成したリストを変数に代入して利用することもできる。

```python
list_num = [num_01, num_02, num_03]
print(list_num)
```

```python
color = ["Red", "Blue", "Green"]
print(color)
```

## 4. if文
プログラムの中でいくつかの条件によって行いたい処理が違う時に条件分岐（if文）を使って行うことができる。  
if文を記述後に条件を記述しその条件がTrueの場合のみ実行される。  

#### 構文の書き方

#### ＜比較演算子＞

■構文例

```python
x = 10
```

```python
if x == 10:
    print("xは10です")
elif x == 5:
    print("xは5です")
else:
    print("xは5でも10でもありません")
```

■構文例

```python
pet = "dog"
```

```python
if pet == "cat":
    print("ペットは猫である")
elif pet in "bird":
    print("ペットは鳥類である")
```

■「〜かつ」、「〜または」の場合

```python
num = 5
```

```python
# 「〜かつ」=and, 「〜または」=orを使う
if num>=0 and num<10:
    print("numは0以上10未満の値である")
```

```python
#不等号の場合、and,orを使わなくてもいい
if 0<=num<10:
    print("numは0以上10未満の値である")
```

## 5. for文
指定回数を繰り返して処理を行いたい時には、for文を使う。  （while文を使う方法もある。)    

#### 構文の書き方 ① 「単純な繰り返し処理」

■構文例

```python
# 以下の処理を5回繰り返す
for i in range(5): 
    #数字を表示する（数字は1づつ増加）
    print(i)           
```

```python
# 始まりの数値を1、増加量を2にしたい場合
for i in range(1, 5, 2):
    print(i)
```

#### 構文の書き方 ② 「データの集まりに対して繰り返し処理」

■構文例: 文字列

```python
#変数charにはHello Worldの文字が順番に代入される
for char in "Hello World": 
    print(char)
```

■構文例: リスト

```python
list_num = [0, 1, 2, 3, 4]
```

```python
 #変数numにはlist_numの要素が添字順に代入される
for num in list_num:
    print(num) 
```

---
#### ＜リストの繰り返し処理のメリット例＞

■問題3の3 for文なし

```python
score = [80, 50, 90]
total = score[0] + score[1] + score[2]  # 長い
print(total)
```

■問題3の3 for文あり

```python
score = [80, 50, 90]
total = 0
for num in score:
    total += num
print(total)
```

## 6. 関数
関数とは「何らかの値を渡すと、その値に応じた何らかの値を返すもの」と定義される。  
関数に渡す何らかの値のことを「引数」、関数から返ってくる何らかの値のことを「戻り値」という。  
Pythonにデフォルトで定義されている関数の代表例にはprint()関数がある。  



Pythonではdef文を用いてユーザーが関数を定義することができる。  
def文の構文の書き方の例は以下の通りである。  

```python
# 仮引数x, yの合計値を戻り値として返す
def add(x, y):
    sum = x + y
    return sum
    
# add関数を呼び出し、戻り値を変数totalに代入
total = add(3, 8) 
print(total)
```
